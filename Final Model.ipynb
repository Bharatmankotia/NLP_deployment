```python
import re
import string
```

```python
 def remove_url(text):
    url=re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r"", text)

def remove_html(text):
    html=re.compile(r'<.*?>')
    return html.sub(r"", text)
```

```python
data_no_duplicates['content']=data_no_duplicates['content'].apply(lambda x: remove_html(x)) # removing html and https content
```

```python
data_no_duplicates.head(50)
```

```python
data_no_duplicates['content'] = data_no_duplicates['content'].str.replace(r'[^\w]+', ' ') # provided space between the words and special characters. Removed punctuations as well
                                                                  # also removed \n from the data     
```

```python
data_no_duplicates.head(50)
```

```python
data_no_duplicates['content'] = data_no_duplicates['content'].str.replace(r'[\d]+', ' ') # removed numbers from the dataset
```

```python
data_no_duplicates.head()
```

```python
data_new=data_no_duplicates[['Class','content']]
```

```python
data_new.head()
```

```python
data_new.isnull().sum() # check for missing values in each column
```

```python
import nltk
stopwords=nltk.corpus.stopwords.words('english')
```

```python
from nltk.stem import WordNetLemmatizer, PorterStemmer
wn=nltk.WordNetLemmatizer()
```

```python
dir(stopwords)
```

```python
def process(text):
    text=[wn.lemmatize(word.lower()) for word in text.split() if word.lower() not in stopwords]
    return " ".join(text)

    
```

```python
data_new['content']= data_new['content'].apply(lambda x: process(x))
```

```python
data_new.head()
```

```python
# without tfidf
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Convert a collection of text documents to a matrix of token counts
cv=CountVectorizer()
```

```python
x=cv.fit_transform(data_new['content'])
```

```python
cv.vocabulary_
```

```python
# Checking for imbalance in the data
import matplotlib.pyplot as plt
count_classes=pd.value_counts(data_new['Class'],sort=True)
count_classes.plot(kind='bar',rot=0)
```

```python
abusive=data_new[data_new['Class']=='Abusive']
non_abusive=data_new[data_new['Class']=='Non Abusive']
```

```python
print(abusive.shape,non_abusive.shape) # our dataset is imbalanced
```

```python
# Building multinomial naive bayes model
from sklearn.naive_bayes import MultinomialNB as MB
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,data_new['Class'],test_size=0.20,random_state=0)
```

```python
from imblearn.over_sampling import SMOTE # importing smote for oversampling to counter imbalanced dataset
```

```python
smote=SMOTE()
```

```python
x_train_smote,y_train_smote=smote.fit_sample(x_train,y_train)
```

```python
from collections import Counter
```

```python
clf=MB()
classifier=clf.fit(x_train_smote,y_train_smote)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred=classifier.predict(x_train_smote)
print(classification_report(y_train_smote,pred))
```

```python
print(confusion_matrix(y_train_smote,pred))
```

```python
print('accuracy:', accuracy_score(y_train_smote,pred)) # accuracy of 0.93
```

```python
pred_test=classifier.predict(x_test)
```

```python
print(confusion_matrix(y_test,pred_test))
```

```python
print('accuracy:', accuracy_score(y_test,pred_test)) # 0.95
```
